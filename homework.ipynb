### Problem 10

The variables `q10` is initialized with 5 steps of the Collatz sequence starting from 15.
Collatz sequences work with the following rule.

```
if the current value x is odd:
    the next value will be 3x+1
else:
    the next value will be x/2
```

Extend `q10` to ten entries following this rule, then plot them as a line chart.
Use the values in `q10` as the y values.
You can use the default values for x in the `plt.plot` call.

````python
# DO NOT CHANGE

q10 = [15, 46, 23, 70, 35]
````

````python
# YOUR CHANGES HERE

import matplotlib.pyplot as plt

# DO NOT CHANGE
q10 = [15, 46, 23, 70, 35]

# YOUR CHANGES HERE

# Extend q10 to 10 entries using the Collatz rules
current_value = q10[-1]
while len(q10) < 10:
    if current_value % 2 != 0:  # Odd rule: 3x + 1
        next_value = 3 * current_value + 1
    else:  # Even rule: x / 2
        next_value = current_value // 2 # Use integer division
    
    q10.append(next_value)
    current_value = next_value

# Plot the values in q10 as the y values. 
# The default x values (0, 1, 2, ..., 9) are automatically used.
plt.plot(q10, marker='o', linestyle='-')
plt.title("First 10 Steps of Collatz Sequence (Starting from 15)")
plt.xlabel("Step Number")
plt.ylabel("Value")

# plt.show()
````

### Problem 15

Use the function `math.comb` to make a bar chart illustrating the binomial distribution for $n=10$.
That is, for integers x from zero to ten, there should be a bar whose height is $10 \choose x$.

````python
# YOUR CHANGES HERE

import math
import matplotlib.pyplot as plt

# 1. Define the parameters
n = 10
x_values = list(range(n + 1)) # x from 0 to 10

# 2. Calculate the binomial coefficients (the heights of the bars)
# y_values = [math.comb(n, x) for x in x_values]
y_values = [math.comb(10, x) for x in x_values]

# 3. Plot the bar chart
plt.bar(x_values, y_values)

# Optional: Add title and labels for clarity
plt.title(f'Binomial Coefficients $\\binom{{10}}{{x}}$')
plt.xlabel('x')
plt.ylabel('Coefficient Value $\\binom{10}{x}$')
plt.xticks(x_values) # Ensure x-axis shows all integer values
plt.grid(axis='y', linestyle='--')
plt.show() 
````

### Problem 17

Plot the factorial function from n=1 to 10 using log scale for better legibility.

Recall that the factorial of $n$, denoted $n!$, is calculated as the product $1 \times 2 \times \ldots \times n$.
The plots that you plot should be $(1, 1!), (2, 2!), (3, 3!), \ldots, (10, 10!)$.
Use [`plt.yscale`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yscale.html) to change your y axis to use log scale.

````python
# YOUR CHANGES HERE

import math
import matplotlib.pyplot as plt

# 1. Define the x-values (n from 1 to 10)
n_values = list(range(1, 11))

# 2. Calculate the y-values (n!)
# Use math.factorial(n) for each n
factorial_values = [math.factorial(n) for n in n_values]

# 3. Plot the data
# Use marker='o' for clarity since it's discrete data
plt.plot(n_values, factorial_values, marker='o')

# 4. Set the y-axis to a logarithmic scale for better legibility
plt.yscale('log')

# Optional: Add title and labels
plt.title('Factorial Function ($n!$) on a Log Scale')
plt.xlabel('n')
plt.ylabel('n! (Log Scale)')
plt.grid(True, which="both", ls="-") # Show grid for both major and minor ticks
plt.xticks(n_values) # Ensure x-axis shows integer values
plt.show()
````

Read the data from `f18.tsv` and parse it as a TSV file.
Divide the data by the "animal" column and in a subplot for each animal, plot the "age" and "weight" columns as the x and y axes respectively with a different subplot for each animal.
Each subplot should be a line chart, but add markers for each point as some animals only have one data point.
Make sure to mention the animal name in the chart titles, and label the axes.

````python
# YOUR CHANGES HERE

import pandas as pd
import matplotlib.pyplot as plt

# 1. Read the TSV file. tab separator='\t' is used for TSV.
try:
    df = pd.read_csv('f18.tsv', sep='\t')
except FileNotFoundError:
    print("Error: f18.tsv not found. Please ensure the file is in the correct directory.")
    # Stop execution if the file isn't found
    # For a solution block, we'll proceed assuming the file exists/will be provided
    pass 

# Determine the number of unique animals to set up the subplots
unique_animals = df['animal'].unique()
num_animals = len(unique_animals)

# Create a figure and a set of subplots
# The layout is (number of animals) rows, 1 column
fig, axes = plt.subplots(nrows=num_animals, ncols=1, figsize=(8, 4 * num_animals))
plt.tight_layout(pad=3.0) # Adjust spacing between subplots

# Ensure 'axes' is an array even for a single subplot
if num_animals == 1:
    axes = [axes]

# 2. Iterate through each unique animal and plot the data
for i, animal_name in enumerate(unique_animals):
    # Filter the DataFrame for the current animal
    animal_data = df[df['animal'] == animal_name]
    
    # Get the correct axis object
    ax = axes[i]
    
    # 3. Plot 'age' (x) vs 'weight' (y)
    # Use mode='lines+markers' (linestyle='-' and marker='o') for line chart with markers
    ax.plot(animal_data['age'], animal_data['weight'], marker='o', linestyle='-')
    
    # 4. Label the axes and set the title
    ax.set_title(f"Weight vs. Age for: {animal_name}")
    ax.set_xlabel("Age")
    ax.set_ylabel("Weight")
    ax.grid(True)

# 5. Display the final plot with all subplots
plt.show()
````

### Problem 20

Read the file `f20.tsv` and parse it as a TSV file.
Three of its columns are color, x, and y.
Plot these x, y values with the specified color using [`plt.scatter`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html).


Hint: you may find it useful to separate the rows by color first, and use a separate call to `plt.scatter` for each color.

````python
# YOUR CHANGES HERE

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Column names based on the problem description
column_names = ['color', 'x', 'y']

# 1. Read the TSV file
# Assuming the first three columns are color, x, and y
df = pd.read_csv('f20.tsv', sep='\t', header=None, names=column_names)

# 2. Get the unique colors (category names) and create a color map
unique_colors = df['color'].unique()

# Map the arbitrary category names to Matplotlib's default color cycle
# C0 is the first color in the cycle, C1 is the second, etc.
# The modulo 10 ensures we cycle through the 10 default colors if there are more categories
color_map = {name: f'C{i % 10}' for i, name in enumerate(unique_colors)}

# 3. Create the scatter plot
plt.figure(figsize=(8, 6))

# Iterate through each unique color category and plot the corresponding points
for color_value in unique_colors:
    # Filter the DataFrame for the current color category
    subset = df[df['color'] == color_value]
    
    # Plot the subset. 
    # Use the VALID color string from the color_map dictionary for the 'c' parameter.
    plt.scatter(
        subset['x'], 
        subset['y'], 
        c=color_map[color_value], # Use the mapped color (e.g., 'C0', 'C1')
        label=color_value,
        s=30 
    )

# Optional: Add labels, title, and legend
plt.title('Scatter Plot of X vs Y, Colored by Category')
plt.xlabel('X-Value')
plt.ylabel('Y-Value')
plt.legend(title='Color Category')
plt.grid(True)
plt.show()
````
